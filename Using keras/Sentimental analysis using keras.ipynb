```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from collections import Counter
```

```python
g = open('/home/lenovo/data files/reviews.txt','r') # input
# data and labels are loaded from file 
reviews = list(map(lambda x:x[:-1],g.readlines()))
g.close()

g = open('/home/lenovo/data files/labels.txt','r') # ground reality
labels = list(map(lambda x:x[:-1].upper(),g.readlines()))
g.close()
```

```python
# TODO: Create set named "vocab" containing all of the words from all of the reviews
vocab = set()
for review in reviews:
    temp = review.split(' ')
    cal = Counter(temp)
    for word in cal:
        vocab.add(word)
vocab_size = len(vocab)
vocab_size
```

```python
# Create a dictionary of words in the vocabulary mapped to index positions
# (to be used in layer_0)
word2index = {}
for i,word in enumerate(vocab):
    word2index[word] = i
    
# display the map of words to indices
word2index
```

```python
# TODO: Create layer_0 matrix with dimensions 1 by vocab_size, initially filled with zeros
layer_0 = np.zeros((1,vocab_size))
```

```python
def update_input_layer(review):
    global layer_0
    # clear out previous state by resetting the layer to be all 0s
    layer_0 *= 0
    
    # TODO: count how many times each word is used in the given review and store the results in layer_0
    temp = review.split(' ')
    cal = Counter(temp)
    for word in cal:
        layer_0[0][word2index[word]] = cal[word]
```

```python
def get_target_for_label(label):
    """Convert a label to `0` or `1`.
    Args:
        label(string) - Either "POSITIVE" or "NEGATIVE".
    Returns:
        `0` or `1`.
    """
    if(label=='POSITIVE'):
        return 1
    else:
        return 0
```

```python
model = Sequential()
model.add(Dense(10,input_shape = (74074,)))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
def train(reviews,labels):
    print("Training : ")
    l = len(reviews)
    correct = 0
    p = 0
    for i in range(l):
        update_input_layer(reviews[i])
        y = get_target_for_label(labels[i])
        x = layer_0
        model.fit(x,np.array([y]), batch_size=1, epochs=1,verbose=0)
        if(model.predict(layer_0)>=0.5):
            p = 1
        else:
            p = 0
        if(p==y):
            correct += 1
        #print("\r"+str(i),end="")
        print("\rProgress : "+str(i*100/l)[:5]+"% "+" Accuracy : "+str(correct*100/(i+1))[:5]+"%",end="")
```

```python
def test(reviews,labels):
    print("Testing : ")
    l = len(reviews)
    correct = 0
    p = 0
    for i in range(l):
        update_input_layer(reviews[i])
        y = get_target_for_label(labels[i])
        x = layer_0
        if(model.predict(x)>=0.5):
            p = 1
        else:
            p = 0
        if(p==y):
            correct += 1
        print("\rProgress : "+str(i*100/l)[:5]+"% "+" Accuracy : "+str(correct*100/(i+1))[:5]+"%",end="")
    
```

```python
train(reviews[:-1000],labels[:-1000])
```

```python
test(reviews[-1000:],labels[-1000:])
```
